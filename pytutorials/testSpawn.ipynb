```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import torch.multiprocessing as mp
from torch.utils.data import DataLoader
from torch.multiprocessing import Process

# Set start method for multiprocessing
mp.set_start_method('spawn', force=True)

# Data transformation
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Download and prepare datasets
trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
testset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Data loaders (parallelized with workers)
trainloader = DataLoader(trainset, batch_size=64, shuffle=True, num_workers=4)
testloader = DataLoader(testset, batch_size=64, shuffle=False, num_workers=4)

# Define the neural network model
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10)

    def forward(self, x):
        x = x.view(-1, 28 * 28)  # Flatten the input
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Training function
def train(rank, num_epochs, device):
    # Create a new model instance for each process
    model = SimpleNN().to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=0.01)

    model.train()  # Set model to training mode
    for epoch in range(num_epochs):
        running_loss = 0.0
        for images, labels in trainloader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
        
        print(f'Rank [{rank}], Epoch [{epoch + 1}/{num_epochs}], Loss: {running_loss / len(trainloader):.4f}')

# Testing function
def test(model, testloader, device):
    model.to(device)  # Move model to correct device
    model.eval()  # Set model to evaluation mode
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in testloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Accuracy of the model: {accuracy:.2f}%')
    return accuracy

# Save the model
def save_model(model, filepath):
    torch.save(model.state_dict(), filepath)
    print(f'Model saved to {filepath}')

# Load the model
def load_model(model, filepath):
    model.load_state_dict(torch.load(filepath))
    model.eval()
    print(f'Model loaded from {filepath}')

# Main function wrapped in __name__ guard
if __name__ == '__main__':
    num_epochs = 5
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Create a list to hold processes
    processes = []
    
    # Create a process for each training task
    for rank in range(4):  # 4 processes (can increase)
        p = Process(target=train, args=(rank, num_epochs, device))
        p.start()
        processes.append(p)

    # Wait for all processes to finish
    for p in processes:
        p.join()

    # Instantiate a new model for testing
    model = SimpleNN().to(device)
    # Load the trained model if needed
    # load_model(model, 'mnist_model.pth')

    # Test the trained model
    test(model, testloader, device)

    # Save the trained model
    save_model(model, 'mnist_model.pth')

    # Visualize some results
    def imshow(img):
        img = img / 2 + 0.5  # Unnormalize
        plt.imshow(img.numpy(), cmap='gray')
        plt.axis('off')

    # Fetch some test data
    dataiter = iter(testloader)
    images, labels = next(dataiter)
    outputs = model(images.to(device))
    _, predicted = torch.max(outputs, 1)

    # Show the first 5 predictions
    plt.figure(figsize=(10, 4))
    for i in range(5):
        plt.subplot(1, 5, i + 1)
        imshow(images[i][0].cpu())
        plt.title(f'Predicted: {predicted[i].item()}')
    plt.show()
```
